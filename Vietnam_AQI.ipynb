```python
import pandas as pd
import matplotlib.pyplot as plt

# load AQI dataset (read_csv)
AQI_data = pd.read_csv("C:/Users/User/Downloads/data_AQI.csv")
AQI_data
```

```python
# Convert to DateTime

AQI_data['Date'] = pd.to_datetime(AQI_data['Date'])
AQI_data['Date']
```

```python
# Filter data for a specific country

vietnam_2024 = AQI_data[(AQI_data['Country'] == 'Vietnam') & (AQI_data['Date'].dt.year == 2024)]
vietnam_2024
```

```python
# Line plot for AQI over time

plt.figure(figsize=( 10, 6))
plt.plot(vietnam_2024['Date'],vietnam_2024['AQI Value'], color = 'blue', label = 'Vietnam AQI')
plt.title('AQI Trend Over Time for Vietnam')
plt.xlabel('Date')
plt.ylabel('Air Quality Index(AQI)')
plt.legend()
plt.grid(True)

plt.show()
```
